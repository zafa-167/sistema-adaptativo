# An√°lise de Dados de Seguran√ßa P√∫blica no Brasil

Este estudo simula dados de criminalidade no Brasil entre os anos de 2023 e 2025, incluindo uma an√°lise por regi√µes para o ano de 2025. O objetivo √© compreender padr√µes de viol√™ncia e propor hip√≥teses sobre causas e poss√≠veis solu√ß√µes, com base em dados inspirados no sistema SINESP.
Neste trabalho, foi similado dados de crimes no Brasil entre 2023 e 2025 para entender como a viol√™ncia tem se comportado nos √∫ltimos anos. O foco foi analisar homic√≠dios, roubos e furtos ‚Äî tanto no pa√≠s todo quanto em diferentes regi√µes.

```python
## 2. Importa√ß√£o de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
## 3. Simula√ß√£o dos dados (exemplo representativo, substitua pelo dataset real do SINESP)
dados = {
    'Ano': [2023, 2024, 2025],
    'Homic√≠dios': [50000, 46000, 42500],
    'Roubos': [120000, 130000, 145000],
    'Furtos': [150000, 160000, 175000]
}

df = pd.DataFrame(dados)
```

```python
## 4. Gr√°fico: Evolu√ß√£o dos crimes no Brasil (2023‚Äì2025)
df_melted = df.melt(id_vars='Ano', var_name='Tipo de Crime', value_name='Ocorr√™ncias')

plt.figure(figsize=(10,6))
sns.lineplot(data=df_melted, x='Ano', y='Ocorr√™ncias', hue='Tipo de Crime', marker='o')
plt.title('Evolu√ß√£o dos Crimes no Brasil (2023-2025)')
plt.ylabel('N√∫mero de Ocorr√™ncias')
plt.xlabel('Ano')
plt.legend(title='Tipo de Crime')
plt.tight_layout()
plt.show()
```

‚úÖ O gr√°fico mostra a tend√™ncia de crimes no Brasil entre 2023 e 2025. Note a queda nos homic√≠dios e o crescimento dos crimes patrimoniais. Homic√≠dios diminu√≠ram: de 50 mil em 2023 para 42,5 mil em 2025.

Roubos e furtos aumentaram: sugerindo que, embora os crimes letais tenham ca√≠do, os patrimoniais (que afetam o dia a dia das pessoas) cresceram.


üí° **Conclus√£o e Recomenda√ß√µes:**

A redu√ß√£o dos homic√≠dios pode estar associada a a√ß√µes mais eficazes de combate ao crime organizado. No entanto, o aumento dos furtos e roubos sugere que fatores como desemprego, urbaniza√ß√£o desordenada e aus√™ncia de pol√≠ticas sociais seguem como desafios.

üîé Recomenda-se:
- Investimentos em preven√ß√£o (educa√ß√£o, gera√ß√£o de emprego)
- Fortalecimento da intelig√™ncia policial
- An√°lise preditiva cont√≠nua com dados reais do SINESP


```python
## 5. Simula√ß√£o de dados por regi√£o em 2025
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
dados_regiao = {
    'Regi√£o': regioes,
    'Homic√≠dios': [8500, 13000, 5000, 9000, 4000],
    'Roubos': [18000, 30000, 10000, 40000, 12000],
    'Furtos': [22000, 35000, 11000, 45000, 13000]
}

df_regiao = pd.DataFrame(dados_regiao)
```

```python
## 6. Gr√°fico: Compara√ß√£o regional (2025)
df_melted_regiao = df_regiao.melt(id_vars='Regi√£o', var_name='Tipo de Crime', value_name='Ocorr√™ncias')

plt.figure(figsize=(12,6))
sns.barplot(data=df_melted_regiao, x='Regi√£o', y='Ocorr√™ncias', hue='Tipo de Crime')
plt.title('Criminalidade por Regi√£o em 2025')
plt.ylabel('N√∫mero de Ocorr√™ncias')
plt.xlabel('Regi√£o')
plt.legend(title='Tipo de Crime')
plt.tight_layout()
plt.show()


```

```python
# Gr√°fico de barras para compara√ß√£o entre regi√µes
df_regiao.set_index('Regi√£o')[['Homic√≠dios', 'Roubos', 'Furtos']].plot(
    kind='bar',
    figsize=(10,6),
    stacked=False
)
plt.title('Distribui√ß√£o de Crimes por Regi√£o (2025)')
plt.ylabel('N√∫mero de Ocorr√™ncias')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

üìä Observa-se que a regi√£o Nordeste lidera em todos os tipos de crime, seguida pelo Sudeste. Isso pode estar ligado √† densidade populacional, desigualdade socioecon√¥mica e infraestrutura de seguran√ßa p√∫blica.


üìå Conclus√£o:


As regi√µes Nordeste e Sudeste concentram os maiores √≠ndices de criminalidade em 2025,
com destaque para roubos e furtos. O Norte tamb√©m apresenta altos √≠ndices de homic√≠dios.
J√° a regi√£o Sul mant√©m os menores registros gerais.
Essas disparidades indicam forte influ√™ncia de fatores estruturais e socioecon√¥micos regionais.

# pergunta 1

1. 1. Qual foi a tend√™ncia observada nos homic√≠dios entre os anos de 2023 e 2025.


```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# GeoDataFrame com as regi√µes do Brasil (simplificadas)
# Aqui vamos associar regi√µes gen√©ricas com um shape b√°sico
regioes = pd.DataFrame({
    'Regi√£o': ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul'],
    'Latitude': [0.5, -7, -12, -20, -27],
    'Longitude': [-63, -42, -55, -46, -51]
})
```

```python
# Mapa de bolhas mostrando a queda proporcional nos homic√≠dios
homicidios_2023 = 50000
homicidios_2025 = 42500
queda_percentual = round((homicidios_2023 - homicidios_2025) / homicidios_2023 * 100, 2)

fig = px.scatter_geo(
    regioes,
    lat='Latitude',
    lon='Longitude',
    text='Regi√£o',
    size=[queda_percentual]*5,
    title=f'Homic√≠dios: Queda de {queda_percentual}% entre 2023 e 2025',
    projection="natural earth"
)
fig.show()

```

Os homic√≠dios apresentaram uma queda consistente, passando de 50.000 em 2023 para 42.500 em 2025. Isso representa uma redu√ß√£o de cerca de 15%, indicando poss√≠vel efic√°cia de pol√≠ticas de repress√£o a crimes violentos.

# Pergunta 2
 2. Como evolu√≠ram os casos de furtos e roubos entre 2023 e 2025, segundo os dados simulados?


```python
import numpy as np

anos = ['2023', '2024', '2025']
roubos = [120000, 135000, 145000]
furtos = [150000, 160000, 175000]

x = np.arange(len(anos))
width = 0.35

fig, ax = plt.subplots()
ax.bar(x - width/2, roubos, width, label='Roubos', color='orange')
ax.bar(x + width/2, furtos, width, label='Furtos', color='blue')

ax.set_ylabel('Ocorr√™ncias')
ax.set_title('Evolu√ß√£o de Roubos e Furtos (2023‚Äì2025)')
ax.set_xticks(x)
ax.set_xticklabels(anos)
ax.legend()
plt.show()
```

Tanto os furtos quanto os roubos aumentaram ao longo do per√≠odo.

Roubos: de 120.000 (2023) para 145.000 (2025)

Furtos: de 150.000 (2023) para 175.000 (2025)
Isso sugere uma tend√™ncia de crescimento nos crimes patrimoniais.
